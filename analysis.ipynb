### 1. Comparing emloyment status of clients in 2019 and 2021 

I intend to compare the proportion of clients that are unemployed and have attended college/university for 2019 and 2021. 

In both df2019 and df2021, the employment status of “Unemployed” is recoded as 3; while the employment status of "Full-time" and "Part-time" is 1 and 2 respectively. The education level of 4 years of college university, BA/BS, some postgraduate study, or more is recoded as 5; while the education level of 1-3 years of college, university, or vocational school is recoded as 4.  

```python
df2019_umemployed_graduate = df2019_filtered[
    (df2019_filtered['EMPLOY'].isin([3])) &
    (df2019_filtered['EDUC'].isin([5]))
]

df2021_umemployed_graduate = df2021_filtered[  
    (df2021_filtered['EMPLOY'].isin([3])) &
    (df2021_filtered['EDUC'].isin([5]))
]

df2019_employed_graduate = df2019_filtered[
    (df2019_filtered['EMPLOY'].isin([1,2])) &
    (df2019_filtered['EDUC'].isin([5]))
]

df2021_employed_graduate = df2021_filtered[  
    (df2021_filtered['EMPLOY'].isin([1,2])) &
    (df2021_filtered['EDUC'].isin([5]))
]

proportion_2019 = len(df2019_umemployed_graduate) / (len(df2019_umemployed_graduate)+len(df2019_employed_graduate))
proportion_2021 = len(df2021_umemployed_graduate) / (len(df2021_umemployed_graduate)+len(df2021_employed_graduate))

print(f"Proportion of college graduates in 2019 unemployed: {proportion_2019:.2%}")
print(f"Proportion of college graduates in 2021 unemployed: {proportion_2021:.2%}")
```

The proportion of unemployed college graduates among all **economically active** college graduduates has increased from 23.41% in 2019 to 36.22% in 2021.

```python
df2019_umemployed_nongrad = df2019_filtered[
    (df2019_filtered['EMPLOY'].isin([3])) &
    (df2019_filtered['EDUC'].isin([4]))
]

df2021_umemployed_nongrad = df2021_filtered[  
    (df2021_filtered['EMPLOY'].isin([3])) &
    (df2021_filtered['EDUC'].isin([4]))
]

df2019_employed_nongrad = df2019_filtered[
    (df2019_filtered['EMPLOY'].isin([1,2])) &
    (df2019_filtered['EDUC'].isin([4]))
]

df2021_employed_nongrad = df2021_filtered[  
    (df2021_filtered['EMPLOY'].isin([1,2])) &
    (df2021_filtered['EDUC'].isin([4]))
]

proportion_2019_nongrad = len(df2019_umemployed_nongrad) / (len(df2019_umemployed_nongrad)+len(df2019_employed_nongrad))
proportion_2021_nongrad = len(df2021_umemployed_nongrad) / (len(df2021_umemployed_nongrad)+len(df2021_employed_nongrad))

print(f"Proportion of unemployed clients who did not finish college in 2019: {proportion_2019_nongrad:.2%}")
print(f"Proportion of unemployed clients who did not finish college in 2021: {proportion_2021_nongrad:.2%}")
```

Similarly, the proportion of unemployed clients who did not finish college with a degree has increased from 37.21% in 2019 to 54.67% in 2021. 

### 2. Strength of association between eductaion level and age of first use of substance.

The age at first use variable `FRSTUSE1` is recoded as the following in both datasets: 

1. 11 years and under;
2. 12-14 years;
3. 15-17 years; 
4. 18-20 years;
5. 21-24 years;
6. 25-29 years;
7. 30 years and older.

The education level variable `EDUC` is recoded as:

1. Less than one school grade, no schooling, nursery school, or kindergarten to Grade 8;
2. Grades 9 to 11;
3. Grades 12 (or GED);
4. 1-3 years of college, university, or vocational school;
5. 4 years of college, university, BA/BS, some postgraduate study, or more. 

I intend to correlate education level with the age of first-time substance use using Spearman’s correlation for admitted clients in year 2019 and 2021 respectively.

```python
from scipy.stats import spearmanr

# Calculate Spearman's correlation for df2019_filtered
correlation_2019, p_value_2019 = spearmanr(df2019_filtered['FRSTUSE1'], df2019_filtered['EDUC'])
print(f"Spearman's correlation for 2019: {correlation_2019}, p-value: {p_value_2019}")

# Calculate Spearman's correlation for df2021_filtered
correlation_2021, p_value_2021 = spearmanr(df2021_filtered['FRSTUSE1'], df2021_filtered['EDUC'])
print(f"Spearman's correlation for 2021: {correlation_2021}, p-value: {p_value_2021}")
```

In 2019, there was a weak but statistically significant positive correlation between education level and age at first use among substance use treatment clients (Spearman's $ρ = 0.0775$, $p < 0.001$). This trend was also observed in 2021, where the correlation remained positive but weaker (Spearman's $ρ = 0.0481$), with the results still being statistically significant ( $p < 5 \times 10^{-97}$). The extremely low p-values is due to the sample size being very large; with large data sets, even very small effects can become highly statistically significant. 

These findings suggest that higher education levels may be slightly associated with later age at first substance use among these clients, though the strength of this association is weak.

### 3. Difference in Age of first Use of Substance for Males and Females

The gender variable “GENDER” is recoded numerically as:
1. Male
2. Female

I intend to perform a Mann-Whitney U test to assess whether age at first use is significantly differnt for males and females in 2019 and 2021 respectively.

```python
from scipy.stats import mannwhitneyu

males_2019 = df2019_filtered[df2019_filtered['GENDER'] == 1]['FRSTUSE1']
females_2019 = df2019_filtered[df2019_filtered['GENDER'] == 2]['FRSTUSE1']

males_2021 = df2021_filtered[df2021_filtered['GENDER'] == 1]['FRSTUSE1']
females_2021 = df2021_filtered[df2021_filtered['GENDER'] == 2]['FRSTUSE1']

u_stat_2019, p_value_2019 = mannwhitneyu(males_2019, females_2019, alternative='two-sided')
u_stat_2021, p_value_2021 = mannwhitneyu(males_2021, females_2021, alternative='two-sided')

print(f"Mann-Whitney U Test for 2019: U-Statistic = {u_stat_2019}, P-value = {p_value_2019}")
print(f"Mann-Whitney U Test for 2021: U-Statistic = {u_stat_2021}, P-value = {p_value_2021}")
```

These results indicate a statistically significant difference in the distributions of age at first use between males and females in 2019. The extremely low p-value (effectively zero) suggests that the observed differences in age distributions are highly unlikely to have occurred by chance.

Similarly, the 2021 results also indicate a statistically significant difference in the distributions of age at first use between genders. The p-value, while not zero, is still exceedingly small, reinforcing the conclusion that the differences observed are not due to random variation but reflect a genuine disparity in the age at first use distributions between males and females.

### 4. Difference in Age of First Substance Use for CLients in Different Living Arrangements

The Living arrangement variable (LIVARAG)
is categorized and recoded numerically as the following: 

1. Homeless
2. Dependent Living
3. Independent Living

I intend to perform a Kruskal-Wallis test to assess whether age at first use is significantly different for different living arrangements for df2019_filtered and df2021_filtered respectively.

```python
from scipy.stats import kruskal

# For df2019_filtered
data_homeless_2019 = df2019_filtered[df2019_filtered['LIVARAG'] == 1]['FRSTUSE1']
data_dependent_2019 = df2019_filtered[df2019_filtered['LIVARAG'] == 2]['FRSTUSE1']
data_independent_2019 = df2019_filtered[df2019_filtered['LIVARAG'] == 3]['FRSTUSE1']

stat_2019, p_2019 = kruskal(data_homeless_2019, data_dependent_2019, data_independent_2019)
print(f"Kruskal-Wallis Test for 2019: Statistic = {stat_2019}, P-value = {p_2019}")

# For df2021_filtered
data_homeless_2021 = df2021_filtered[df2021_filtered['LIVARAG'] == 1]['FRSTUSE1']
data_dependent_2021 = df2021_filtered[df2021_filtered['LIVARAG'] == 2]['FRSTUSE1']
data_independent_2021 = df2021_filtered[df2021_filtered['LIVARAG'] == 3]['FRSTUSE1']

stat_2021, p_2021 = kruskal(data_homeless_2021, data_dependent_2021, data_independent_2021)
print(f"Kruskal-Wallis Test for 2021: Statistic = {stat_2021}, P-value = {p_2021}")
```

These results indicate statistically significant differences in the ages at first use among individuals in different living arrangements (Homeless, Dependent Living, Independent Living) in 2019. The very low p-value suggests that these differences are highly unlikely to have occurred by chance, implying that living conditions may play a significant role in influencing the age at first substance use.

Similarly, in 2021, the test results show statistically significant differences between the living arrangement groups regarding age at first use. The higher test statistic and even lower p-value compared to 2019 reinforce the impact of living arrangements on the initiation of substance use. The evidence strongly suggests that the age at first use varies significantly across different living scenarios.

### 5. Association between source of income/support and the age at first substance use

I intend to perform Chi-square test of independence to examine the association between the source of income/support and the age at first substance use for the years 2019 and 2021.  

The source of Income/support (“PRIMINC”) is a categorical variable recoded as the following: 
1. Wages/salary
2. Public assistance
3. Retirement/pension, disability
4. Other
5. None

```python
from scipy.stats import chi2_contingency

# Prepare the data for 2019
df2019_clean = df2019_filtered[df2019_filtered['PRIMINC'] != -9]

# Create a contingency table for 2019
table_2019 = pd.crosstab(df2019_clean['PRIMINC'], df2019_clean['FRSTUSE1'])

# Prepare the data for 2021
df2021_clean = df2021_filtered[df2021_filtered['PRIMINC'] != -9]

# Create a contingency table for 2021
table_2021 = pd.crosstab(df2021_clean['PRIMINC'], df2021_clean['FRSTUSE1'])
```

```python
# Perform the Chi-square test for 2019
chi2_stat_2019, p_value_2019, dof_2019, expected_2019 = chi2_contingency(table_2019)
print("Chi-square Test for 2019:")
print("Chi2 Statistic:", chi2_stat_2019, "P-value:", p_value_2019)

# Perform the Chi-square test for 2021
chi2_stat_2021, p_value_2021, dof_2021, expected_2021 = chi2_contingency(table_2021)
print("\nChi-square Test for 2021:")
print("Chi2 Statistic:", chi2_stat_2021, "P-value:", p_value_2021)
```

The findings from 2019 demonstrate a statistically significant association between the source of income/support and the age at first substance use, indicating that different income sources are associated with when individuals first begin using substances.

Similarly, the 2021 results also show a statistically significant association, confirming the findings from 2019. This suggests a consistent pattern over the two years studied, highlighting the influence of economic factors on substance use behaviors.

The significant Chi-square statistics for both years indicate that economic conditions, as measured by sources of income, are strongly associated with the age at first substance use. These results suggest that interventions aiming to address substance use may need to consider the economic backgrounds of the individuals to tailor more effective prevention and treatment strategies.

### 5. Association between source of income/support and the age at first substance use


```python
# Assuming df2019_clean is your DataFrame and already filtered out '-9' for PRIMINC
income_labels = {
    1: '1. Wages/salary',
    2: '2. Public assistance',
    3: '3. Retirement/pension, disability',
    4: '4. Other',
    5: '5. No source of income'
}
age_labels = {
    1: '1-11 years and under',
    2: '12-14 years',
    3: '15-17 years',
    4: '18-20 years',
    5: '21-24 years',
    6: '25-29 years',
    7: '30 years and older'
}

df2019_clean['PRIMINC_Label'] = df2019_clean['PRIMINC'].map(income_labels)
df2019_clean['FRSTUSE1'] = df2019_clean['FRSTUSE1'].map(age_labels)
```

```python
# Create a crosstabulation of the data
ct = pd.crosstab(df2019_clean['PRIMINC_Label'], df2019_clean['FRSTUSE1'])

# Normalize the crosstab to get proportions instead of counts
ct_normalized = ct.div(ct.sum(axis=1), axis=0)
ct_normalized = ct_normalized.reset_index()
ct_melted = pd.melt(ct_normalized, id_vars=['PRIMINC_Label'], var_name='AgeAtFirstUse', value_name='Proportion')
```

```python
plt.figure(figsize=(10, 8))
chart = sns.barplot(x='PRIMINC_Label', y='Proportion', hue='AgeAtFirstUse', data=ct_melted, palette='Set2')

chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
chart.set_title('Normalized Clustered Bar Chart of Source of Income and Age at First Use for 2019')
plt.legend(title='Age at First Use', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

Note that the age range of 15 to 17 are predominantly higher across all income types, which indicates that individuals within this age group generally begin using substances more frequently compared to other age groups, regardless of their source of income.
